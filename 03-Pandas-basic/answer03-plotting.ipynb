### 导入必要的库

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
```

### 载入数据

```python
path2 = "./Euro 2012 stats TEAM.csv"      # Euro2012_stats.csv
```

```python
euro12 = pd.read_csv(path2)
euro12
```

### 问题1：用直方图展示各队进球数

```python
euro12.Goals.plot(kind='bar')
```

### 问题2：散点图分析射正数`Shots on target`与破门数`Goals`之间关系

```python
plt.plot(euro12['Shots on target'], euro12['Goals'], 'ro')
plt.xlabel('Shots on target')
plt.ylabel('Goals')
```

### 将数据集中的列Team, Yellow Cards和Red Cards单独存为一个名叫discipline的数据框

```python
discipline = euro12[['Team', 'Yellow Cards', 'Red Cards']]
discipline
```

### 问题3：画饼图展示红黄牌占比

```python
plt.pie(
    [discipline['Yellow Cards'].sum(), discipline['Red Cards'].sum()],
    labels=['Yellow Cards', 'Red Cards'],
)

plt.axis('equal')
```
