```python
import numpy as np
import pandas as pd
```

```python
titanic = pd.read_csv('./titanic.csv')
```

### 问题1：男/女幸存者与遇难者分别为多少人？

```python
s2s = titanic.pivot_table(index=['Sex'], columns=['Survived'], values=['PassengerId'], aggfunc='count')
s2s
```

```python
s2s = pd.crosstab(titanic.Sex, titanic.Survived)
s2s
```

### 问题2：男/女各年龄段幸存者与遇难者分别为多少人？

```python
as2s = pd.crosstab([titanic.Sex, titanic.Age], titanic.Survived, margins=True)
pd.concat([as2s.head(), as2s.tail()])
```

```python
bins = np.arange(0, 110, 10)
titanic['AgeBin'] = pd.cut(titanic.Age, bins)
titanic.head()
```

```python
as2s = pd.crosstab([titanic.Sex, titanic.AgeBin], titanic.Survived, margins=True)
as2s
```
