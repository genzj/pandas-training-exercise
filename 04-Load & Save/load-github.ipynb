```python
import numpy as np
import pandas as pd
import requests
```

### 1. 使用requests库下载数据

```python
req = requests.get('https://api.github.com/users/genzj/repos?per_page=100')
raw = req.json()
raw
```

### 2. 将数据转换成DataFrame

```python
data1 = pd.DataFrame(raw)
data1.head()
```

### 3. 使用`full_name`作为索引

```python
data=data1.set_index('full_name')
data.head()
```

### 4. 观察数据基本形态
可以直接得出：仓库数量（id-count）、star/fork最大值

```python
data.describe()
```

### 5. 画出仓库大小直方图

```python
%matplotlib inline
data['size'].sort_values().plot(kind='bar', sort_columns=True)
```
