## 收益率计算

```python
# 基本原理
(aapl['Adj Close'].iloc[1:].to_numpy() / aapl['Adj Close'].iloc[:-1].to_numpy()) - 1
```

```python
# 使用pandas计算收益率
aapl['Adj Close'] / aapl['Adj Close'].shift(1) - 1
```

```python
# 该计算比较常用，因此有内置方法
aapl.pct_change()['Adj Close']
```

```python
# 结合resample计算月收益率
aapl.resample('M').mean().pct_change()['Adj Close']
```

```python
ax = aapl.resample('M').mean().pct_change()['Adj Close'].plot()
aapl.pct_change()['Adj Close'].plot(ax=ax)
```

```python
# 对于收益率，更常见的绘图方式是分布图

# aapl.resample('M').mean().pct_change()['Adj Close'].plot(kind='hist', bins=50, label='M')
aapl.pct_change()['Adj Close'].plot(kind='hist', bins=50)
plt.legend()
```

```python
ax = aapl.pct_change()['Adj Close'].plot(kind='hist', bins=50, density=True)
aapl.resample('M').mean().pct_change()['Adj Close'].plot(kind='hist', bins=50, label='M', density=True, ax=ax, alpha=0.8)
plt.legend()
```

```python
# 累计收益率是收益倍数的累积乘积
daily_change = aapl.pct_change()
cum_daily_return = (1 + daily_change).cumprod()
ax = daily_change['Adj Close'].plot(color='orange', alpha=0.5)
cum_daily_return['Adj Close'].plot(label='cum', ax=ax)
```

```python
# 两个尺度不同的量，绘制到一起时比例关系不匹配，可以分两个纵轴画
daily_change = aapl.pct_change()
cum_daily_return = (1 + daily_change).cumprod()
ax = daily_change['Adj Close'].plot(color='orange', alpha=0.5)
ax.set_ylabel('日收益率')


ax2 = ax.twinx()
ax2.set_ylabel('累积收益率')
cum_daily_return['Adj Close'].plot(label='cum', ax=ax2)

legends = ax.get_legend_handles_labels()[0] + ax2.get_legend_handles_labels()[0]
labels = ax.get_legend_handles_labels()[1] + ax2.get_legend_handles_labels()[1]
ax.legend(legends, labels, loc='upper left')
```
